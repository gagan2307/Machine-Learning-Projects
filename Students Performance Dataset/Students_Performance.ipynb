# Spliting up data set

```python
df = df.drop('StudentID',axis=1)
```

```python
df = df.drop('Gender',axis=1)
```

```python
df = df.drop('Ethnicity',axis=1)
```

```python
df = df.drop('GradeClass',axis=1)
```

```python
X = df.iloc[:,:-2]
y = df['GPA']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

# Training Linear

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr_pre = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, lr_pre)
r2 = r2_score(y_test, lr_pre)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

# Training Ridge

```python
rid = Ridge()
rid.fit(X_train,y_train)
```

```python
rid_pre = rid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, rid_pre)
r2 = r2_score(y_test, rid_pre)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

# Training Lasso

```python
las = Lasso()
las.fit(X_train,y_train)
```

```python
las_pre = las.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, las_pre)
r2 = r2_score(y_test, las_pre)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

# Training Decesion Tree

```python
dtree = DecisionTreeRegressor()
```

```python
dtree.fit(X_train,y_train)
```

```python
dtree_pre = dtree.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, dtree_pre)
r2 = r2_score(y_test, dtree_pre)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

# Training Random Forest

```python
rf = RandomForestRegressor(n_estimators=500)
rf.fit(X_train,y_train)
```

```python
rf_pre = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, rf_pre)
r2 = r2_score(y_test, rf_pre)
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

# CONCLUSION
1. Linear and Ridge performed the best with 95% r2 score
2. Random Forest perform the second best with 92% r2 score
3. all other performed least then them at around 83% r2 score

# Predicting Using linear

```python
def gclass (prediction):
    
    if prediction >= 3.5:
        return "GradeClass 0"
    elif 3.0 <= prediction < 3.5:
        return "GradeClass 1"
    elif 2.5 <= prediction < 3.0:
        return "GradeClass 2"
    elif 2 <= prediction < 2.5:
        return "GradeClass 3"
    elif prediction < 2.0:
        return "GradeClass 4"
```
